**Question no 24**

```python
# From the question number 24
Population_mean= 270
Sample_size = 18
Sample_mean = 260
sample_Std_dev= 90
alpha= 0.05
```

T-Value

```python
std_error= sample_Std_dev/np.sqrt(Sample_size)
t_value= (Sample_mean-Population_mean)/std_error
t_value
```

```python
p_value= stats.t.cdf(t_value, df=17)
p_value
```

```python
if (p_value>alpha):
  print("The company's claim is not acepted")
else:
  print("The company's claim is  acepted")
```

Since the confidence interval is not given we assume it as 95% and alpha = 0.05. T-value is 0.471 and p-value is 0.321. We consider this as degree of freedom is 17. Here P > alpha. The null hypothesis is accepted and the average bulb last for 270 days.
